```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

from sklearn.datasets import load_breast_cancer
```

```python
# Load data and split into feature and target variables
data = load_breast_cancer()

data_df = pd.DataFrame(data.data, columns=data.feature_names)
target_df = pd.DataFrame(data.target, columns=['target'])
combined_df = pd.concat([data_df, target_df], axis=1)
```

```python
# View first 5 rows of the data
data_df.head()
```

```python
# How frequently does the positive target occur?
target_df['target'].value_counts()
```

```python
# Generate summary statistics for the data
data_df.describe()
```

```python
# Create a pairplot for the first few features
sns.pairplot(combined_df, hue='target', vars=['mean radius', 'mean texture', 'mean perimeter'])
```

```python
# Create a correlation coefficeint heatmap
sns.heatmap(data_df.corr())
```

```python
# Create a boxplot for mean radius by target from target_df type
sns.boxplot(x='target', y='mean radius', data=combined_df)
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data_df, target_df, test_size=0.2, random_state=42)
```

```python
# Build and train logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# Evaluate the model
y_pred = logreg.predict(X_test)
```

```python
# Generate a confusion matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('True')
```

```python
# Generate a classification report
print(classification_report(y_test, y_pred))
```

```python
# Extract coefficients from trained model
coefficients = logreg.coef_[0]
```

```python
# Normalize the coefficients by the standard deviation
std_devs = data_df.std().values
normalized_coefficients = coefficients / std_devs
```

```python
# Sort feature names and coefficients by absolute value of coefficients
sorted_indices = np.argsort(np.abs(normalized_coefficients))[::-1]
sorted_feature_names = data_df.columns[sorted_indices]
sorted_coefficients = normalized_coefficients[sorted_indices]
```

```python
# Visualize feature importances
plt.figure(figsize=(10, 5))
plt.barh(y=sorted_feature_names, width=sorted_coefficients)
plt.xlabel('Normalized Coefficient')
plt.ylabel('Feature')
plt.title('Feature Importances')
```
